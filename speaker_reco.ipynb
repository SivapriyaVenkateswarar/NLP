# Data Loading and extracting



```python
import librosa
import librosa.display # Import librosa.display explicitly
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio
import scipy
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os

data_path = '/content/drive/MyDrive/AI /NLP/archive (20)/16000_pcm_speeches'
speaker_folders = os.listdir(data_path)
x=[]
y=[]
for speaker in speaker_folders:
    speaker_path = os.path.join(data_path, speaker)
    if os.path.isdir(speaker_path):

        audio_files = os.listdir(speaker_path)
        for file in audio_files:
            file_path = os.path.join(speaker_path, file)
            if os.path.isfile(file_path):

                audio,sr=librosa.load(file_path, sr=8000)
                mfcc=librosa.feature.mfcc(y=audio,sr=sr,n_mfcc=13)
                mean_mfcc=np.mean(mfcc,axis=1)

                x.append(mean_mfcc)
                y.append(speaker)

X = np.array(x)
y = np.array(y)

print(f"Extracted MFCCs shape: {X.shape}")
print(f"Labels shape: {y.shape}")

```

# DATA SPLITTING:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

# MODEL TRAINING:

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=1000)
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```

# ACCURACY MEASUREMENT:

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cm=confusion_matrix(y_test, y_predict)
cm
```

```python
print(classification_report(y_test, y_predict))
score=round(accuracy_score(y_test, y_predict),4)*100
print('Accuracy score: {}% '.format(score))
```
